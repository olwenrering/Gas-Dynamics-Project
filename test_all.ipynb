```python
from pycesam import *
from pycesam.gui import *
import astropy.units as u
from astropy.constants import R_jup, R_sun, mu0, M_sun, G
import matplotlib.pyplot as plt
import numpy as np
%matplotlib widget
from ipywidgets import interact
```

```python
paths = ["./","./1.1MJ/","./1.2MJ/","./1.3MJ/","./1.4MJ/","./1.5MJ/","./1.6MJ/","./1.7MJ/","./1.8MJ/","./1.9MJ/",
         "./2MJ/"]

mdl = []
mdl_plot = []

for i in range(len(paths)):
    print(i)
    if i ==0:
        print("il veut pas avec le path")
        mdl.append(CModel('1MJ_R'))
        mdl_plot.append(CModelGUI('1MJ_R'))
    elif i >=5:
        mdl.append(CModel(paths[i]+'2MJ'))
        mdl_plot.append(CModelGUI(paths[i]+'2MJ'))
    else:
        mdl.append(CModel(paths[i]+'1MJ'))
        mdl_plot.append(CModelGUI(paths[i]+'1MJ'))

mdlt = CModel('1MJ_R')
mdl_plott = CModelGUI('1MJ_R')
```

```python
plt.figure()
loop = len(mdl)
for i in range(loop):
    mdl_plot[i].plot_cz(r_or_m='r')
plt.show()
```

```python
end_list=[]
Omega_list=[]
R_tot_list=[]
for i in range(loop):

    end = len(mdl[i].glob)-1
    end_list.append(end)
    Omega = mdl[i].params.w_rot*1e3
    Omega_list.append(Omega)
    R_tot = mdl[i].var[end][0][0]/1e2
    R_tot_list.append(R_tot)
    
    print(f"total radius of planet = {R_tot/(R_jup.value)} Rjup")
    print(f"total mass of planet = {mdl[i].params.mtot} Msun")
    print(f"total angular velocity of planet = {Omega/1e3} km/s")
    print("\n")
```

```python
B_time = []
plt.figure()

@interact(i=(0, len(mdl[0].var)-1, 1)) 
def plot_Elsasser_over_time(i=0):

    for k in range(loop):
        B_time.append([])
        T = mdl[k].var[i][2]
        sigma = 10**7*T**(3/2)
        c_light = 3*10**10
        eta = c_light**2/(4*np.pi*sigma)
        rho = mdl[k].var[i][4]
        mdl[k].calc_vconv('osc', i=i)
        #Rm = mdl.vconv*mdl.lmix/eta
        mask = mask_list[k]
        rho_avg = np.mean(rho[mask])
        qc_0 = mdl[k].Fconv[mask][-1]
        dr = mdl[k].var[i][0][mask]
        mask_2 = np.where(mdl[k].Fconv[mask] > 0)
        R_tot = mdl[k].var[i][0][0]
        r = ((mdl[k].var[i][0])/(R_tot))
        index = np.where(np.abs(T - T[-1]/2.78) < 200)
        H_T_obs = (r[index]*R_tot)[0]
        #print(f"H_T_obs =  {H_T_obs}")

        
        V = (4/3)*np.pi*(R_tot)**3
        dF = (mdl[k].Fconv[mask][mask_2]*mdl[k].lmix[mask][mask_2]/(qc_0*H_T_obs))**(2/3)*(rho[mask][mask_2]/rho_avg)**(1/3)*4*np.pi*dr[mask_2]**2
        mask_3 = np.where(dF > 0)
        F = np.trapezoid(dF[mask_3][::-1], dr[mask_2][mask_3][::-1]) / V 
        #F= np.trapz(dF[mask_3][::-1], dr[mask_2][mask_3][::-1]) / V
        print(f"F = {F}")


        
        
        B = np.sqrt(c*F*rho_avg**(1/3)*qc_0**(2/3))
        B_time[k].append(B)
        Omega = mdl[k].var[i][15]
        Elsasser = B**2/(2*eta*rho*Omega)
        #Elsasser_prime = Elsasser*mdl.lmix/(0.6*np.sqrt(Rm))
        
        print(f"Generated B = {B} Gauss")
        
        
        plt.plot(r, Elsasser, color=color_names[k],label=paths[k])
    #plt.plot(r, Elsasser_prime, label='$\Lambda$ prime')
    plt.plot(r, 1*np.ones(len(Elsasser)), label=r'$\Lambda$ = 1')
    plt.xlabel(r'r/R$_p$')
    plt.ylabel(r'$\Lambda$ (Elsasser nb)')
    plt.yscale('log')
    plt.legend()
    plt.show()

```
