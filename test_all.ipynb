```python
c_light = 3*10**10

def compute_B(model, i):
    T = model.var[i][2]
    rho = model.var[i][4]
    model.calc_vconv('osc', i=i)
    mask = np.where(~np.isnan(model.vconv))
    rho_avg = np.mean(rho[mask])
    qc_0 = model.Fconv[mask][-1]
    dr = model.var[i][0][mask]
    mask_2 = np.where(model.Fconv[mask] > 0)
    R_tot = model.var[i][0][0]
    #print(R_tot)
    r = ((model.var[i][0])/(R_tot))
    index = np.where(np.abs(T - T[-1]/2.78) < 100)
    if len(index[0]) == 0:
        index = np.where(np.abs(T - T[-1]/2.78) < 300)
    H_T_obs = (r[index]*R_tot)[0]
    #print(H_T_obs)
    V = (4/3)*np.pi*(R_tot)**3
    dF = (model.Fconv[mask][mask_2]*model.lmix[mask][mask_2]/(qc_0*H_T_obs))**(2/3)*(rho[mask][mask_2]/rho_avg)**(1/3)*4*np.pi*dr[mask_2]**2
    mask_3 = np.where(dF > 0)
    F = np.trapezoid(dF[mask_3][::-1], dr[mask_2][mask_3][::-1]) / V 
    #F= np.trapz(dF[mask_3][::-1], dr[mask_2][mask_3][::-1]) / V
    #print(f"F = {F}")
    B = np.sqrt(c*F*rho_avg**(1/3)*qc_0**(2/3))
    return B

print(len(mdl[0][0].glob)-1)
print(compute_B(mdl[0][0], len(mdl[0][0].glob)-1))
print(np.where(np.abs(np.array(mdl[2][0].glob)[:,10]-500)<=10))
print(np.shape(np.array(mdl[0][0].var)))
```

```python
plt.close('all')

def plot_Elsasser_over_time(i=47, k=0):
    plt.close('all')
    plt.figure()
    for j in range(len(mdl[k])):
        T = mdl[k][j].var[i][2]
        sigma = 10**7*T**(3/2)
        eta = c_light**2/(4*np.pi*sigma)
        rho = mdl[k][j].var[i][4]
        B = compute_B(mdl[k][j], i)
        Omega = mdl[k][j].var[i][15]
        Elsasser = B**2/(2*eta*rho*Omega)
        #print(f"Generated B = {B} Gauss")
        R_tot = mdl[k][j].var[i][0][0]
        r = ((mdl[k][j].var[i][0])/(R_tot))
        age = mdl[k][j].glob[i][10]
        plt.plot(r, Elsasser,label=f"{mdl[k][j].params.w_rot} km/s")
        #plt.plot(r, Elsasser_prime, label='$\Lambda$ prime')
        if j==0:
            plt.plot(r, 1*np.ones(len(Elsasser)), label=r'$\Lambda$ = 1')
        plt.title(f'{mdl[k][j].params.mtot*1e3:.1f} MJ At t = {age:.2f} Myr')
    plt.xlabel(r'r/R$_p$')
    plt.ylabel(r'$\Lambda$ (Elsasser nb)')
    plt.yscale('log')
    plt.legend()
    plt.show()

interact(plot_Elsasser_over_time, i=(0, len(mdl[-1][0].var)-1, 1), k=(0, loop-1, 1))
```
