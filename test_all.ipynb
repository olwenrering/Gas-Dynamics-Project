```python
import os
from pathlib import Path
import shutil
import traceback
import numpy as np

models = []
models_plot = []

masses = np.arange(1, 2.1, 0.1)
base_root = Path("/home/localuser/Desktop/M2/Gas_Dynamics/Project/test_planet/1MJ/rotating")

# store the original cwd so we can always restore to a known place
orig_cwd = Path.cwd()

for i, mass in enumerate(masses):
    models.append([])
    models_plot.append([])

    mass_dir = base_root / f"{mass:.1f}MJ"
    
    if not mass_dir.exists():
        print(f"Warning: mass directory {mass_dir} does not exist  skipping")
        continue

    

    # loop over rotation speeds
    for w in range(5, 11):          # range(5,20,1) is same as range(5,20)
        try:
            # prepare the working directory for this rotation
            work_dir = mass_dir / str(int(w))
            """if i != 0:
                mass_dir_before = base_root / f"{masses[i-1]:.1f}MJ" / str(int(w))
                shutil.copy(mass_dir_before / "1MJ_R.run", work_dir / "1MJ_R.run")
            # if not present, copy from the "5" template
            if w != 5:
                template = mass_dir / "5"
                if work_dir.exists():
                    shutil.rmtree(work_dir)
                shutil.copytree(template, work_dir)"""
            # change to the working directory; always restore in the finally below
            prev_cwd = Path.cwd()
            os.chdir(work_dir)

            # create models and append them (use appended element rather than indexing by w-5)
            m = CModel('1MJ_R')
            mgui = CModelGUI('1MJ_R')
            models[i].append(m)
            models_plot[i].append(mgui)

            # set params and launch if needed
            # note: convert units consistently: your earlier code used mtot = masses[i]*1e-3
            """desired_mtot = mass * 1e-3
            if (np.abs(m.params.mtot-desired_mtot) >= 1e-4) or (m.params.w_rot != w):
                m.params.mtot = desired_mtot
                m.params.w_rot = w
                print(f"launching simulation at {m.params.mtot} Msun and w = {m.params.w_rot} km/s")

                try:
                    m()   # run simulation
                except Exception as e_sim:
                    # print traceback and continue to next w
                    print(f"Simulation failed for mass={mass:.1f}MJ w={w}: {e_sim}")
                    traceback.print_exc()
                    # optionally: continue to next w (do not break)
                    continue"""

        except Exception as e_outer:
            # top-level errors (e.g., directory missing, copytree failed)
            print(f"Error preparing/handling mass={mass:.1f}MJ w={w}: {e_outer}")
            traceback.print_exc()
            # continue to next w (or break if you prefer)
            continue

        finally:
            # always restore previous working directory (if prev_cwd is set)
            try:
                os.chdir(prev_cwd)
            except Exception:
                # if something went wrong and prev_cwd is not defined or missing,
                # restore to original base_root for safety
                os.chdir(orig_cwd)

    # end for w
# end for masses

# restore original cwd at the end (defensive)
os.chdir(orig_cwd)
print("All done  cwd restored.")
```

```python
B_time = []
plt.figure()

@interact(i=(0, len(mdl[0].var)-1, 1)) 
def plot_Elsasser_over_time(i=0):

    for k in range(loop):
        B_time.append([])
        T = mdl[k].var[i][2]
        sigma = 10**7*T**(3/2)
        c_light = 3*10**10
        eta = c_light**2/(4*np.pi*sigma)
        rho = mdl[k].var[i][4]
        mdl[k].calc_vconv('osc', i=i)
        #Rm = mdl.vconv*mdl.lmix/eta
        mask = mask_list[k]
        rho_avg = np.mean(rho[mask])
        qc_0 = mdl[k].Fconv[mask][-1]
        dr = mdl[k].var[i][0][mask]
        mask_2 = np.where(mdl[k].Fconv[mask] > 0)
        R_tot = mdl[k].var[i][0][0]
        r = ((mdl[k].var[i][0])/(R_tot))
        index = np.where(np.abs(T - T[-1]/2.78) < 200)
        H_T_obs = (r[index]*R_tot)[0]
        #print(f"H_T_obs =  {H_T_obs}")

        
        V = (4/3)*np.pi*(R_tot)**3
        dF = (mdl[k].Fconv[mask][mask_2]*mdl[k].lmix[mask][mask_2]/(qc_0*H_T_obs))**(2/3)*(rho[mask][mask_2]/rho_avg)**(1/3)*4*np.pi*dr[mask_2]**2
        mask_3 = np.where(dF > 0)
        F = np.trapezoid(dF[mask_3][::-1], dr[mask_2][mask_3][::-1]) / V 
        #F= np.trapz(dF[mask_3][::-1], dr[mask_2][mask_3][::-1]) / V
        print(f"F = {F}")


        
        
        B = np.sqrt(c*F*rho_avg**(1/3)*qc_0**(2/3))
        B_time[k].append(B)
        Omega = mdl[k].var[i][15]
        Elsasser = B**2/(2*eta*rho*Omega)
        #Elsasser_prime = Elsasser*mdl.lmix/(0.6*np.sqrt(Rm))
        
        print(f"Generated B = {B} Gauss")
        
        
        plt.plot(r, Elsasser, color=color_names[k],label=paths[k])
    #plt.plot(r, Elsasser_prime, label='$\Lambda$ prime')
    plt.plot(r, 1*np.ones(len(Elsasser)), label=r'$\Lambda$ = 1')
    plt.xlabel(r'r/R$_p$')
    plt.ylabel(r'$\Lambda$ (Elsasser nb)')
    plt.yscale('log')
    plt.legend()
    plt.show()

```
